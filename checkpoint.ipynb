```python
#create 1d array and insert the values 1,2,3,,5 and access the following elements:1st element,2nd element,range of elements from 1st to 3rd,range of elements from 2nd element,range of elements from index=-2 

import numpy as np

# Create the array

arr = np.array([1, 2, 3, 5])

# 1st element (index 0)
first_element = arr[0]

# 2nd element (index 1)
second_element = arr[1]

# Range of elements from 1st to 3rd (index 0 to 2, inclusive of start, exclusive of stop)
range_1_to_3 = arr[0:3]

# Range of elements from 2nd element to the end (index 1 to end)
range_from_2nd = arr[1:]

# Range of elements from index = -2 to the end (negative index counts from the end)
range_from_neg2 = arr[-2:]

print("1st element:", first_element)
print("2nd element:", second_element)
print("Range from 1st to 3rd element:", range_1_to_3)
print("Range from 2nd element to end:", range_from_2nd)
print("Range from index -2 to end:", range_from_neg2)
```

```python
#craete a 1d array make it within order  by axis wise.

import numpy as np

# Create a 1D array with unordered elements
arr = np.array([5, 3, 9, 1, 7])

# Sort the array to order it axis-wise (in ascending order)
arr_sorted = np.sort(arr)

print("Original array:", arr)
print("Sorted array:", arr_sorted)
```

```python
#read 1d array and insert the values 1,9,2,5,3 and check whether the elements in array are greater than 5.and print those elements as separate list(use boolean masking)

import numpy as np

# Create 1D NumPy array with given values
arr = np.array([1, 9, 2, 5, 3])

# Boolean mask for elements greater than 5
mask = arr > 5

# Filter elements based on mask
filtered_elements = arr[mask]

print("Original array:", arr)
print("Elements greater than 5:", filtered_elements.tolist())  # convert to list for separate list output
```
