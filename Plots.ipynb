```python
m=np.array([[1,4,2],
          [4,7,4],
          [2,0,9]])
print(m)
```

```python
print('Matrix Transpose:\n',m.transpose(),'\n')
print('Matrix determinant:\n',np.linalg.det(m),'\n')
m_inv=np.linalg.inv(m)
print('Matrix inverse:\n',m_inv,'\n')
iden_m=np.dot(m,m_inv)
iden_m=np.round(np.abs(iden_m),0)
print('Product of matrix and its inverse:\n',iden_m)
```

```python
eigen_vals,eigen_vecs=np.linalg.eig(m)
print('Eigen Values',eigen_vals,'\n')
print('Eigen vectors:\n',eigen_vecs)
```

```python
U,S,VT=np.linalg.svd(m)
print('Getting SVD outputs:\n')
print('U:\n',U,'\n')
print('S:\n',S,'\n')
print('VT:\n',VT,'\n')
```

```python
#get data
import scipy as sp
import numpy as np
nums=np.random.randint(1,20,size=(1,15))[0]#generates a 1-diamentional numpy array containing 15 random
print('Data:',nums)
```

```python
#get descriptive state
print('Mean:', sp.mean(nums))
print('Median:', sp.median(nums))
print('Mode:', sp.stats.mode(nums))
print('Standard Deviation:', sp.std(nums))
print('Variance:', sp.var(nums))
print('Skew:',sp.stats.skew(nums))
print('Kurtosis:',sp.stats.kurtosis(nums))
```

```python
#write a code to generate 100 random lottery tickets and pick 2 lucky tickets from it as a winner
#the lottery number must be 10 digits long
#all 100 ticket numbers must be unique
#generate a random list of 1000 numbers usng
import random
lottery_tickets_list = []

print("Creating 100 random lottery tickets...")
for _ in range(100):
    ticket = random.randint(1000000000, 9999999999)
    lottery_tickets_list.append(ticket)

# ðŸŽ‰ Randomly select 2 lucky winners
winners = random.sample(lottery_tickets_list, 2)

print("Lucky 2 lottery tickets are:", winners)
```

```python
#write a code to generate a random password which meets the following condition
#password length must be 10 characters long
#it must contain
import random
import string

def randomPassword():
    # Define the source of characters
    randomSource = string.ascii_letters + string.digits + string.punctuation

    # Ensure a strong password structure
    password = [
        random.choice(string.ascii_uppercase),  # At least one uppercase
        random.choice(string.ascii_lowercase),  # At least one lowercase
        random.choice(string.digits),           # At least one digit
        random.choice(string.punctuation)       # At least one special character
    ]

    # Add more random characters to reach desired length (e.g., 10)
    password += random.choices(randomSource, k=6)

    # Shuffle the resulting list to avoid predictable patterns
    random.SystemRandom().shuffle(password)

    # Join the list into a string
    return ''.join(password)

# Generate and print the password
print("Password is", randomPassword())
```

```python
fig,ax=plt.subplots()
x_index=3
colors=['blue','red','green']
for label,color in zip(range(len(iris.target_names)),colors):
    ax.hist(iris.data[iris.target==label,x_index],
           label=iris.target_names[label],
           color=color)
ax.set_xlabel(iris.feature_names[x_index])
ax.legend(loc='upper right')
plt.show()
```

```python
fig,ax=plt.subplots()
x_index=3
y_index=0
colors=['blue','red','green']
for label,color in zip(range(len(iris.target_names)),colors):
    ax.scatter(iris.data[iris.target==label,x_index],
               iris.data[iris.target==label,y_index],
           label=iris.target_names[label],
           c=color)
ax.set_xlabel(iris.feature_names[x_index])
ax.set_ylabel(iris.feature_names[y_index])
ax.legend(loc='upper right')
plt.show()
```
