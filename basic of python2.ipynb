```python
#method5
#creating dataframe using zip() function

import pandas as pd

#list1
students=['Alan','Vivian','Alistar','Jade']

#list2
age=[24,26,32,29]

#usig zip to merge two lists
list_of_tuples=list(zip(students,age))

#list(zip(students,age)) will return

#converting the lists ofr tuples to pandas dataframe
df=pd.DataFrame(list_of_tuples,columns=['students','age'])
display(df)
```

```python
import pandas as pd

Name=['Alice','Bob','Charlie','David','Karl','Amile','Daminam']
age=[25,30,22,28,19,22,27]
scores=[85,70,90,80,75,77,82]
city=['newyork','Los Angeles','Frankfurt','London','Oslo','Brcelona','Frankfurt']
list_of_tuples=list(zip(Name,age,scores,city))
df=pd.DataFrame(list_of_tuples,columns=['Name','age','scores','city'])
display(df)
```

```python
# 1.selection by column names(single or list)
selected_df_1=df['Name'] #selecting a single column by name
selected_df_2=df[['Name','scores']] #ng a list of column names
selected_df_1
selected_df_2

```

```python
# 2.Selection by positional indexing
selected_df_3=df[0:2] #selecting first 2 columns(zero-based indexing)
selected_df_4=df.iloc[:,1:3] #selecting columns from index 1(inclusive) to
selected_df_3
selected_df_4

```

```python
# 3.Selection with .Loc(label-based)
selected_df_6=df.loc[:,['Name','city']] #selecting columns from 'name' and'city'
(selected_df_6)
```

```python
#sorting rows within pandas dataframe
import pandas as pd

raw_data={"first_name":["Sheldon","Raj","Lenord","Howard","Amy"],
         "last_name":["Copper","Koothrappali","Hofstadter","Wolwitz","Fowler"],
                     "age":[42,38,36,41,35],
                     "Comedy_Score":[9,7,8,8,5],
         "Rating_Score":[25,25,49,62,70]}

df=pd.DataFrame(raw_data,columns=["first_name","last_name","age","Comedy_Score","Rating_Score"])
print(df)
```

```python
print(df.sort_values(by="age",ascending=0))
print();
print(df.sort_values(by="age",ascending=1))
print();
print(df.sort_values(by=["age","Comedy_Score"]))
print();
```

```python
#groupBy on a Single Column(Group)
group_g=df.groupby(["first_name"])
print(group_g)
print(group_g.size())
```

```python
#How to  find the  largest and Lowest value in a pandas dataframe
print("Index of largest value:");print(df['Comedy_Score'].idxmax())
print("Index of lowest value:");print(df['Comedy_Score'].idxmin())
```
