```python
#sorting rows with pandas dataframe
import pandas as pd
raw_data={'first_name':['sheldon','raj','leonard','fathima','any'],
         'last_name':['has','kumar','peter','rashidha','sina'],
         'age':[32,12,43,12,44],
         'comedy_score':[9,5,3,7,6],
         'rating_score':[25,25,43,23,65]}

df=pd.DataFrame(raw_data,columns=['first_name','last_name','age','comedy_score','rating_score'])
print(df)

```

```python
print(df.sort_values(by='age',ascending=0))
print()
```
